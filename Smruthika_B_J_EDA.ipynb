```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
transactions = pd.read_csv("Transactions.csv")

# Display initial rows of each dataset
print(customers.head())
print(products.head())
print(transactions.head())
```

```python
# Check data info and summaries
customers.info()
products.info()
transactions.info()

# Check for missing values
print(customers.isnull().sum())
print(products.isnull().sum())
print(transactions.isnull().sum())

# Check for duplicates
print(customers.duplicated().sum())
print(products.duplicated().sum())
print(transactions.duplicated().sum())
```

```python
# Convert dates to datetime format
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])

# Drop duplicates (if any)
customers = customers.drop_duplicates()
products = products.drop_duplicates()
transactions = transactions.drop_duplicates()

# Fill missing values (example)
customers.fillna({'Region': 'Unknown'}, inplace=True)
```

```python
# Region distribution
region_counts = customers['Region'].value_counts()
sns.barplot(x=region_counts.index, y=region_counts.values)
plt.title('Customer Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Top product categories
category_counts = products['Category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Top Product Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Example transactions DataFrame (Replace this with your actual dataset loading code)
data = {
    'TransactionDate': ['2025-01-01', '2025-01-15', '2025-02-01', '2025-02-10', '2025-03-05'],
    'TotalValue': [100, 200, 300, 150, 400]
}
transactions = pd.DataFrame(data)

# Ensure 'TransactionDate' is in datetime format
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])

# Group by month and calculate the total sales
sales_trends = transactions.groupby(transactions['TransactionDate'].dt.to_period('M'))['TotalValue'].sum()

# Plot the sales trend over time
sales_trends.plot(kind='line', marker='o')
plt.title('Sales Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales (USD)')
plt.grid(True)
plt.show()
```
